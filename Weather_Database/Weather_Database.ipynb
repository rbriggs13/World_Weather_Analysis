```python
#import dependencies
import numpy as np
from citipy import citipy
from config import weather_api_key
import requests
import json
import pandas as pd
```

```python
#generate 2,000 random latitudes and longitudes
lats = np.random.uniform(-90, 90, size=2000)
lngs = np.random.uniform(-180, 180, size=2000)
lat_lngs = zip(lats, lngs)
lat_lngs
```

```python
#add the latitudes and longitudes to a list
coordinates = list(lat_lngs)
```

```python
#creating a list of cities closest to each randomly generated pair
cities = []

for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    #if the city isn't on the list add it to the list
    if city not in cities:
        cities.append(city)
len(cities)
```

```python
#create a data frame and lists to store the data
cities_df = pd.DataFrame()

cities_df['City'] = ""
cities_df['Country'] = ""
cities_df['Lat'] = ""
cities_df['Lng'] = ""
cities_df['Max Temp'] = ""
cities_df['Humidity'] = ""
cities_df['Cloudiness'] = ""
cities_df['Wind Speed'] = ""
cities_df['Current Description'] = ""




cities_df.head()
```

```python
#perform an API call for all of the cities
base_url = "https://api.openweathermap.org/data/2.5/weather?"
index_count = 0

for city in cities:
    try:
        params = {"q": city, "appid": weather_api_key, 'units': 'imperial'}

        weather_data = requests.get(base_url, params=params)
        weather_data = weather_data.json()
    
        cities_df.loc[index_count, 'City'] = weather_data['name']
        cities_df.loc[index_count, 'Country'] = weather_data['sys']['country']
        cities_df.loc[index_count, 'Lat'] = weather_data['coord']['lat']
        cities_df.loc[index_count, 'Lng'] = weather_data['coord']['lon']
        cities_df.loc[index_count, 'Max Temp'] = weather_data['main']['temp_max']
        cities_df.loc[index_count, 'Humidity'] = weather_data['main']['humidity']
        cities_df.loc[index_count, 'Cloudiness'] = weather_data['clouds']['all']
        cities_df.loc[index_count, 'Wind Speed'] = weather_data['wind']['speed']
        cities_df.loc[index_count, 'Current Description'] = weather_data['weather'][0]['description']
    
        index_count += 1
        print('City Found')
        
    except:
        print('City Not Found')
        pass
    

    
```
